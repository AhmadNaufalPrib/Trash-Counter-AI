```python
# 1. Install library yang dibutuhkan
!pip install ultralytics

# 2. Hubungkan ke Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 3. Import library YOLO
from ultralytics import YOLO

# 4. Load model dasar (YOLOv8 Nano - ringan untuk pemula)
model = YOLO('yolov8n.pt')

# 5. Mulai Pelatihan (Training)
# Pastikan path ke data.yaml sudah benar sesuai lokasi di Drive Anda
results = model.train(
    data='/content/drive/MyDrive/Dataset_Sampah/data.yaml',
    epochs=50,      # AI akan belajar sebanyak 50 putaran
    imgsz=640,      # Ukuran gambar standar
    device=0        # Menggunakan GPU
)
```

```python
# Ganti 'path/to/your/image.jpg' dengan lokasi foto sampah baru
model = YOLO('/content/runs/detect/train3/weights/best.pt') # Muat hasil training tadi

# Jalankan deteksi
results = model.predict(source='/content/drive/MyDrive/Dataset_Sampah/images/test/m-_140226155933-552.jpg', save=True)

# Tampilkan hasil
for result in results:
    print(f"Jumlah sampah terdeteksi: {len(result.boxes)}")
```
